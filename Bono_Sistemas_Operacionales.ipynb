```python
import numpy as np
import matplotlib.pyplot as plot
```

```python
nump=3 #Numero de procesos
numax=10 #Rango del numero aleatorio a generar a partir del 0
numat=5000 #Numero de tablas a comparar

prom1fcfs=np.zeros(numat)
prom2fcfs=np.zeros(numat)
prom1sjf=np.zeros(numat)
prom2sjf=np.zeros(numat)
for w in range (0,numat):
  array = np.random.randint(numax, size=(nump, 2))

  waitime = np.zeros((nump, 1))
  usado = np.zeros((nump,1))

  time=numax+1
  for x in range (0,nump):
      if array[x][1] < time:
        time = array[x][1];
  z=0
  final = 0
  while(final==0):
    menor=numax+1
    for x in range (0,nump):
      if array[x][1] < menor and usado[x][0]==0:
        menor = array[x][1];
        z = x
    usado[z][0]=1
    if time<array[z][1]:
      time=array[z][1]
    waitime[z][0]= time
    time= time + array[z][0]
    final=1
    for a in range (0,nump):
      if usado[a][0]==0:
        final=0

  fcfs = np.zeros((nump,4))
  promwt=0
  promtat=0
  for x in range(0,nump):
    fcfs[x][0]=array[x][0]
    fcfs[x][1]=array[x][1]
    fcfs[x][2]=waitime[x][0]-fcfs[x][1]
    promwt=promwt+fcfs[x][2]
    fcfs[x][3]=fcfs[x][0]+fcfs[x][2]
    promtat=promtat+fcfs[x][3]

  promwt=promwt/nump
  prom1fcfs[w]=promwt
  promtat=promtat/nump
  prom2fcfs[w]=promtat
  print (fcfs)
  print ("Mean of Waiting Time: ",promwt)
  print ("Mean of Turn Around Time: ",promtat)

  #----------------------------------------------------------
  waitime = np.zeros((nump, 1))
  usado = np.zeros((nump,1))
  time=0
  final=0
  while(final==0):
    menor=numax+1
    for x in range (0,nump):
      if array[x][0] < menor and usado[x][0]==0:
        menor = array[x][0];
        z = x
    usado[z][0]=1
    waitime[z][0]= time
    time= time + array[z][0]
    final=1
    for a in range (0,nump):
      if usado[a][0]==0:
        final=0

  sjf = np.zeros((nump,4))
  promwt=0
  promtat=0
  for x in range(0,nump):
    sjf[x][0]=array[x][0]
    sjf[x][1]=array[x][1]
    sjf[x][2]=waitime[x][0]
    promwt=promwt+sjf[x][2]
    sjf[x][3]=sjf[x][0]+sjf[x][2]
    promtat=promtat+sjf[x][3]

  promwt=promwt/nump
  prom1sjf[w]=promwt
  promtat=promtat/nump
  prom2sjf[w]=promtat
  print (sjf)
  print ("Mean of Waiting Time: ",promwt)
  print ("Mean of Turn Around Time: ",promtat)

prom1fcfs=sorted(prom1fcfs)
prom2fcfs=sorted(prom2fcfs)
prom1sjf=sorted(prom1sjf)
prom2sjf=sorted(prom2sjf)
```

```python
from scipy.stats import norm 
import statistics
```

```python
n, bins, patches=plot.hist(prom1fcfs) 
plot.xlabel("Values") 
plot.ylabel("Frequency") 
plot.title("Histogram") 
plot.show()

n, bins, patches=plot.hist(prom1sjf)
plot.xlabel("Values")
plot.ylabel("Frequency")
plot.title("Histogram")
plot.show()
```

```python

n, bins, patches=plot.hist(prom2fcfs)
plot.xlabel("Values")
plot.ylabel("Frequency")
plot.title("Histogram")
plot.show()

n, bins, patches=plot.hist(prom2sjf)
plot.xlabel("Values")
plot.ylabel("Frequency")
plot.title("Histogram")
plot.show()
```

Distribucion Normal del Waiting Time:

```python
m1fcfs = statistics.mean(prom1fcfs) 
sd1fcfs = statistics.stdev(prom1fcfs) 
plot.plot(prom1fcfs, norm.pdf(prom1fcfs, m1fcfs, sd1fcfs),color='darkred')

m1sjf = statistics.mean(prom1sjf) 
sd1sjf = statistics.stdev(prom1sjf) 
plot.plot(prom1sjf, norm.pdf(prom1sjf, m1sjf, sd1sjf),color='darkblue')
```

Distribucion Normal del TurnAround Time:

```python
m2fcfs = statistics.mean(prom2fcfs) 
sd2fcfs = statistics.stdev(prom2fcfs) 
plot.plot(prom2fcfs, norm.pdf(prom2fcfs, m2fcfs, sd2fcfs),color='darkred')

m2sjf = statistics.mean(prom2sjf) 
sd2sjf = statistics.stdev(prom2sjf) 
plot.plot(prom2sjf, norm.pdf(prom2sjf, m2sjf, sd2sjf),color='darkblue')
```
